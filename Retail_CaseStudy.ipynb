```python
Product_hierarchy = pd.read_csv(r'C:\Users\ADMIN\Downloads\prod_cat_info.csv')
```

```python
Transaction = pd.read_csv(r'C:\Users\ADMIN\Downloads\Transactions.csv')
Transaction
```

```python
Customer_Trans=pd.merge(left=Customer,
                        right=Transaction,
                        left_on='customer_Id',
                        right_on='cust_id',
                        how='inner',
                       )
```

```python
Customer_Trans.shape
```

```python
Customer_Final=pd.merge(left=Customer_Trans,
                        right=Product_hierarchy,
                        left_on='prod_cat_code',
                        right_on='prod_cat_code',
                        how='inner'
                       )
```

```python

import numpy as np
Data_min = Customer_Final['total_amt'].min()
Data_max = Customer_Final['total_amt'].max()
Data_q1  = np.percentile(Customer_Final.total_amt,25)
median  = np.percentile(Customer_Final.total_amt,50)
Data_q3  = np.percentile(Customer_Final.total_amt,75)
print('Min = ',Data_min)
print('Max = ',Data_max)
print('Median = ',median)
print('Q1 = ',Data_q1)
print('Q3 = ',Data_q3)
```
